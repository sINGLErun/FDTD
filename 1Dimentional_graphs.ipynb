```python
import matplotlib.pyplot as plt
%matplotlib inline

#
# const int SRC_POS = 100;
# const int DET_NUM = 5;
# const int DET_POS[DET_NUM] = {0, SRC_POS, SIZE_X/2-10, SIZE_X/2+10, SIZE_X-1};
#


DET_NUM = 5;

f = [open("det._values" + str(i) + ".txt", 'r') for i in range(DET_NUM)]

Ez = [[float(v) for v in f[i]] for i in range(DET_NUM)]
time = range(int(Ez[0][0]))

fig = plt.figure(figsize=(8, 8))

ax1 = fig.add_subplot(511, ylim = (-0.6, 0.6) )
ax1.plot(time[1200:1999], Ez[0][1201:2000], color = 'b')

ax2 = fig.add_subplot(512, ylim = (-0.6, 1))
ax2.plot(time[1200:1999], Ez[1][1201:2000], color = 'r')

ax3 = fig.add_subplot(513, ylim = (-0.6, 1))
ax3.plot(time[1200:1999], Ez[2][1201:2000], color = 'purple')

ax4 = fig.add_subplot(514, ylim = (-0.6, 1))
ax4.plot(time[1200:1999], Ez[3][1201:2000], color = 'g')

ax5 = fig.add_subplot(515, ylim = (-0.6, 1))
ax5.plot(time[1200:1999], Ez[4][1201:2000], color = 'orange')

for i in range(DET_NUM):
    f[i].close
```
