```python
import matplotlib.pyplot as plt
%matplotlib inline

f = open("det._values.txt", 'r')

Ez = [float(v) for v in f]
time = range(int(Ez[0]))

fig, ax = plt.subplots(figsize=(8, 8))
ax.plot(time[:], Ez[1:])

f.close()
```
